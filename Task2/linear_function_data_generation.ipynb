```python
from random import uniform
from matplotlib import pyplot
```

```python
a = 5
b = 10

min_delta = -100
max_delta = 100

min_x = -100
max_x = 100
points_count = 100
```

```python
def function(x):
    return a * x + b + uniform(min_delta, max_delta)
```

```python
x_values = [uniform(min_x, max_x) for _ in range(points_count)]
y_values = [function(x) for x in x_values]


pyplot.scatter(x_values, y_values, s=4)
pyplot.plot([min_x, max_x], [min_x * a + b, max_x * a + b], color='r')
```

```python
with open("linear_regression_data.csv", 'w') as output:
    output.write("x, y\n")
    for i in range(len(x_values)):
        x = x_values[i]
        y = y_values[i]
        output.write(f"{x},{y}\n")
```
